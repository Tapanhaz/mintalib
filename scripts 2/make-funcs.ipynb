# Notebook to generate funcs module

```python
import project

from mintalib import core
```

```python
def core_functions():
    """ list of core functions """

    return [
        k for k, v in vars(core).items()
        if k.isupper() and callable(v)
    ]
    


print(core_functions())
```

```python
def make_funcs():
    buffer = "# Do not edit. File was generated by make-funcs!\n\n"

    names = core_functions()
    names = "\n    " + ",\n    ".join(names) + "\n"

    buffer += f"from .core import ({names})\n"

    return buffer

code = make_funcs()
print(code)
```

```python
import importlib.util

def new_module(name):
    spec = importlib.util.spec_from_loader(name, None)
    return importlib.util.module_from_spec(spec)

code = make_funcs()

funcs = new_module("mintalib.funcs")

exec(code, funcs.__dict__)

dir(funcs)
```

```python
outfile = project.pkgdir / "funcs.py"

print(f"Updating {outfile.name} ...")

outfile.write_text(code)
```
