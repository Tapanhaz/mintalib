# Bollinger Bands Function

```python
import matplotlib as mpl

from mintalib.utils import sample_prices

import mintalib.functions as ta

mpl.rc('figure', figsize=(12, 9), facecolor='w')
```

```python
prices = sample_prices(260)
prices = prices.join(ta.bbands(prices))
prices.tail()
```

```python
prices.plot(
    y=['close', 'upperband', 'middleband', 'lowerband'],
    style=['b', 'y--', 'k--', 'y--']
);
```
