```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

from mintalib.core import calc_bbands
from mintalib.utils import sample_prices

mpl.rc('figure', figsize=(12, 9), facecolor='w')
pd.set_option('display.max_rows', None)
```

```python
prices = sample_prices(260)
prices = prices.join(calc_bbands(prices))
prices
```

```python
prices.plot(y=['close', 'upperband', 'middleband', 'lowerband'], style=['b', 'y--', 'k--', 'y--']);
```
