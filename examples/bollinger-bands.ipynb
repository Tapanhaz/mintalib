# Bollinger Bands Function

```python
import matplotlib as mpl

from mintalib.functions import BBANDS
from mintalib.utils import random_prices

mpl.rc("figure", figsize=(12, 9))
```

```python
prices = random_prices(260)
prices = prices.join(BBANDS(prices))
prices.tail()
```

```python
prices.plot(
    y=["close", "upperband", "middleband", "lowerband"],
    style=["b", "y--", "k--", "y--"],
);
```
